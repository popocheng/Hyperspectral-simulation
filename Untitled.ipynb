```python
import numpy as np
import matplotlib.pyplot as plt

# 生成植被目标
def generate_vegetation_target(width, height, num_channels):
    vegetation = np.zeros((height, width, num_channels))
    # 模拟植被在可见光波段的光谱特性，这里简单地生成一个在绿光波段具有较高反射率的植被
    green_reflectance = np.random.uniform(0.5, 1, size=(height, width))
    vegetation[:, :, 10:15] = green_reflectance[..., np.newaxis]
    return vegetation

# 生成海洋生物目标
def generate_marine_life_target(width, height, num_channels):
    marine_life = np.zeros((height, width, num_channels))
    # 模拟海洋生物在不同波长下的光谱特性，这里简单地生成随机的反射率
    marine_life[:, :, :] = np.random.rand(height, width, num_channels)
    return marine_life

# 生成潜艇目标
def generate_submarine_target(width, height, num_channels):
    submarine = np.zeros((height, width, num_channels))
    # 模拟潜艇在不同波长下的光学特性，这里简单地生成随机的反射率
    submarine[:, :, :] = np.random.rand(height, width, num_channels)
    return submarine

# 合成包含三种不同目标的海洋高光谱图像
def generate_hyperspectral_image_with_targets(width, height, num_channels):
    # 生成海洋背景
    ocean_background = np.random.rand(height, width, num_channels)
    # 生成植被目标
    vegetation_target = generate_vegetation_target(width, height, num_channels)
    # 生成海洋生物目标
    marine_life_target = generate_marine_life_target(width, height, num_channels)
    # 生成潜艇目标
    submarine_target = generate_submarine_target(width, height, num_channels)
    # 将目标叠加到海洋背景中
    hyperspectral_image = ocean_background + vegetation_target + marine_life_target + submarine_target
    return hyperspectral_image

# 可视化海洋高光谱图像
def visualize_hyperspectral_image(wavelengths, image):
    num_channels = image.shape[2]
    fig, axes = plt.subplots(1, num_channels, figsize=(15, 3))
    for i in range(num_channels):
        axes[i].imshow(image[:, :, i], cmap='gray')
        axes[i].set_title(f"Wavelength: {wavelengths[i]} nm")
        axes[i].axis('off')
    plt.tight_layout()
    plt.show()

# 主函数
def main():
    width = 100  # 图像宽度
    height = 100  # 图像高度
    num_channels = 20  # 波长通道数
    wavelengths = np.linspace(400, 700, num_channels)  # 波长范围400-700nm
    hyperspectral_image = generate_hyperspectral_image_with_targets(width, height, num_channels)
    visualize_hyperspectral_image(wavelengths, hyperspectral_image)

if __name__ == "__main__":
    main()
```
